```python
sent1 = 'A man, a plan, a canal: Panama'
print(sent1)
```

```python
def clean_string(s):
    """Removes non-alphanumeric characters and converts to lowercase."""
    return "".join(ch.lower() for ch in s if ch.isalnum())
clean_sent = clean_string(sent1)
print(clean_sent)
```

```python
# 1. Palindrome Using String Slicing
def palindrome_slice(s):
    cleaned = clean_string(s)
    # [::-1] creates a reversed copy of the string
    return cleaned == cleaned[::-1]
print(palindrome_slice(sent1))  # True
```

```python
# 2. Palindrome Using Two-Pointer Loop
def palindrome_loop(s):
    cleaned = clean_string(s)
    left, right = 0, len(cleaned) - 1
    while left < right:
        if cleaned[left] != cleaned[right]:
            return False
        left += 1
        right -= 1
    return True
print(palindrome_loop(sent1))  # True
```

```python
# 3. Palindrome Using 'reversed()' Built-in
def palindrome_reversed(s):
    cleaned = clean_string(s)
    # reversed() returns an iterator, so we join it back into a string
    return cleaned == "".join(reversed(cleaned))
print(palindrome_reversed(sent1))  # True
```

```python
# 4. Palindrome Using Recursion
def palindrome_recursive(s):
    cleaned = clean_string(s)
    
    def check(sub):
        # Base case: if 0 or 1 chars left, it's a palindrome
        if len(sub) <= 1:
            return True
        # Compare first and last characters
        if sub[0] != sub[-1]:
            return False
        # Recursive step: check the inner substring
        return check(sub[1:-1])
        
    return check(cleaned)
print(palindrome_recursive(sent1))  # True
```

```python
print(f"Original: {sent1}")
print(f"Cleaned:  {clean_string(sent1)}")
print("-" * 30)
print(f"Method 1 (Slicing):   {palindrome_slice(sent1)}")
print(f"Method 2 (Looping):   {palindrome_loop(sent1)}")
print(f"Method 3 (Reversed):  {palindrome_reversed(sent1)}")
print(f"Method 4 (Recursive): {palindrome_recursive(sent1)}")
```
